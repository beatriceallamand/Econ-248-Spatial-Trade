## Baseline Model: Armington + Rosen Roback

CES preferences. Location $j$ chooses $\{q_{ij}\}_{i\in S}$ to
$$
W_j = \max_{q_{ij}} \left( \sum_{i \in S} q_{ij}^{\frac{\sigma - 1}{\sigma}} \right)^{\frac{\sigma}{\sigma - 1}} U_j
$$
$$\text{s.t.} \sum_{i \in S} q_{ij} \cdot p_{ij} \leq X_j$$

Yields demand
$$q_{ij}=a_{ij}p_{ij}^{-\sigma} X_j P_j^{\sigma -1}$$

Replacing $X_j = P_j V_j$ from last problem set:
$$q_{ij}=a_{ij}\left(\frac{p_{ij}}{P_j}\right)^{-\sigma} V_j$$

Where $P_j$ is the price index:

$$P_j=\left( \sum_{i\in S} a_{ij} p_{ij}^{1 - \sigma} \right)^{\frac{1}{1 - \sigma}} $$


## 1. Armington (Inner Loop)

$$p_i=\frac{w_i}{A_i}$$
$$p_{ij}=\tau_{ij}\frac{w_i}{A_i}$$

Substitute $p_{ij}$ into demand:
$$q_{ij}=a_{ij}p_{ij}^{-\sigma} X_j P_j^{\sigma -1}= a_{ij} \tau_{ij}^{-\sigma} \left( \frac{w_i}{A_i}\right)^{-\sigma} X_j P_j^{\sigma -1}$$

Note $X_{ij}=p_{ij} q_{ij}$
$$X_{ij} = p_{ij}  q_{ij}= a_{ij} \tau_{ij}^{1-\sigma} \left( \frac{w_i}{A_i}\right)^{1-\sigma} X_j P_j^{\sigma -1}$$


Define
$\lambda_{ij}$ as fraction of income of $j$ spent in good $i$. 

$$ \lambda_{ij}=\frac{X_{ij}}{X_j} = a_{ij} \tau_{ij}^{1-\sigma} \left( \frac{w_i}{A_i}\right)^{1-\sigma} P_j^{\sigma -1}$$

Then
$$\underbrace{Y_j = L_j w_j}_{\text{income in $j$}} = \underbrace{\sum_{i \in S} \lambda_{ji} w_i L_i = \sum_{i \in S} X_{ji}}_{\text{what all locations spend in $j$}}$$

In matrix form

$$ Y = \lambda Y$$

Where
$ Y = \begin{bmatrix}
Y_1  \\
Y_2 \\
\vdots \\
Y_S
\end{bmatrix} = \begin{bmatrix}
w_1 L_1  \\
w_2 L_2 \\
\vdots \\
w_S L_S
\end{bmatrix}  $ and $\lambda= \begin{bmatrix}
\lambda_{11} & \lambda_{12} & \dots & \lambda_{1S} \\

**DETOUR: Problem 1, Problem Set 2024. The Basic Armington Model without Labor Mobility.**
  
There is a discrete set $S$ of locations. Each location produces its own distinct variety of a differentiated good. Within each location there is perfect competition among firms. All firms produce with the same labor-only technology:

$$
y = A_i l_i
$$

where $A_i$ is a location-specific productivity term. Workers in each location inelastically provide one unit of labor to the local labor market and spend all their income on consuming the differentiated varieties of the good. They have CES preferences.

There are iceberg trade costs $\tau_{ij} \geq 0$ between locations, so that $\tau_{ij}$ units have to be shipped from $i$ of a given good for 1 unit to arrive in destination $j$.

Assume:  
$$S = 10, \quad a_{ij} = 1 \ \forall i,j \in S, \quad \sigma = 2$$  
$$L_i = 1 \ \forall i, \quad A_i = 1, 2, 3, \ldots, 10$$

Solve for the equilibrium of this model when  

$$
\tau_{ii} = 1, \quad \tau_{ij} = 2 \ \forall i \ne j
$$

Then re-compute the equilibrium when  

$$
\tau_{ij} = 1 \ \forall i,j
$$

Graph the welfare gains from the trade cost reduction in each region against its underlying productivity. Verify that the formula for welfare gains from trade cost changes we derived in class holds in your model. Solution hidden but here.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set parameters
S = 10  # number of regions
sigma = 2.0  # CES elasticity
a_ij = 1.0  # taste parameter
L = np.ones(S)  # labor supply in each region is 1
A = np.arange(1, S + 1)  # productivity A_i = 1, ..., 10. In this version we have no amenities!


# Define trade cost matrices
tau_high = np.full((S, S), 2.0) # tau_ij = 2 if i  j
np.fill_diagonal(tau_high, 1.0) # tau_ii = 1

tau_low = np.ones((S, S)) # free trade: tau_ij = 1 for all i, j


# Fixed point solver on Y = lambda(Y) * Y
def solve_Y_lambdaY(tau, A, sigma, max_iter=1000, tol=1e-6):
    """
    Solves for equilibrium income vector Y using the fixed point Y = lambda(Y) * Y,
    where Y_i = income = w_i * L_i, and L_i = 1 => Y_i = w_i.
    lambda_ji = share of income in i spent on goods from j
    """
    S = len(A)
    Y = np.copy(A)  # Initial guess: assume autarky wages => Y_i = w_i = A_i

    for _ in range(max_iter):
        # Step 1: Compute price index P_j = [sum_i (p_ij)^(1 - sigma)]^{1/(1 - sigma)}
        P = np.array([
            sum((tau[i, j] * Y[i] / A[i]) ** (1 - sigma) for i in range(S)) ** (1 / (1 - sigma))
            for j in range(S)
        ])

        # Step 2: Compute lambda_ij = (p_ij / P_j)^(1 - sigma)
        lambda_matrix = np.zeros((S, S))
        for j in range(S):  # Spending region
            for i in range(S):  # Producing region
                lambda_matrix[i, j] = ( (tau[i, j] * Y[i] / A[i]) / P[j]) ** (1 - sigma)


        # Step 3: Update income Y using fixed point Y =  Y
        Y_new = lambda_matrix.T @ Y
        if np.max(np.abs(Y_new - Y)) < tol:
            break
        Y = Y_new.copy()
        
        welfare = Y / P  # real income (welfare)

    return Y, lambda_matrix, P, welfare

# Solve the model under both trade regimes
Y_high, lambda_high, P_high, welfare_high = solve_Y_lambdaY(tau_high, A, sigma)
Y_low, lambda_low, P_low, welfare_low = solve_Y_lambdaY(tau_low, A, sigma)

# Prepare results
df_results = pd.DataFrame({
    "Location": np.arange(1, S + 1),
    "Productivity A_i": A,
    "Y (High Trade Costs)": Y_high,
    "Y (Free Trade)": Y_low,
    "P (low)": P_low,
    "P (high)": P_high,
    "Welfare (high)": welfare_high,
    "Welfare (low)": welfare_low,
})
df_results

# Plot results
df_results.set_index("Location")[["Y (High Trade Costs)", "Y (Free Trade)"]].plot(
    kind="bar", figsize=(10, 5), title="Equilibrium Income (Y = Y)"
)
plt.ylabel("Income Y = w_i L_i")
plt.grid(True)
plt.tight_layout()
plt.show()

# Compute welfare ratios
welfare_ratio = welfare_low / welfare_high

# Plot
plt.figure(figsize=(8, 5))
plt.plot(A, welfare_ratio, marker='o', color='steelblue', linewidth=1)
plt.xlabel("Productivity $A_i$", fontsize=12)
plt.ylabel("Welfare Ratio (Free Trade / High Trade Cost)", fontsize=12)
plt.title("Welfare Gains from Trade Liberalization", fontsize=14)

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_visible(True)

plt.grid(False)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()

#
```

## Problem 1. The Basic Armington Model with Free Labor Mobility.


Consider the exact same Armington model as in the last problem set. Now we introduce one difference: workers can move freely. This introduces an additional equilibrium condition: the real wage (it's actually utility)  
$$
W_i = \frac{w_i}{P_i} U_i
$$  
has to be equal across locations in equilibrium. Solve for the equilibrium with  
$$
\tau_{ij} = 2
$$  
and for the one with  
$$
\tau_{ij} = 1
$$  
separately. Graph the change in local population between the two scenarios against location productivity. The local amenities, $U_i$, are mirror images of the productivity parameter $A_i$. The region with $A_i = 1$ has $U_i = 10$, the one with $A_i = 2$ has $U_i = 9$ and so on.  

*Hint: relative to the old code you should now add an "outer loop."* First solve for the wage on the inner loop holding population constant. Then compute the real wage in each location. In the outer loop, you then add some population to locations which have a real wage higher than the median region, and take workers away from regions with a real wage below the median. In this way you iterate between updating wages holding populations fixed, and updating populations holding wages fixed.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML

# Set parameters
S = 10  # number of regions
sigma = 2.0  # CES elasticity
a_ij = 1.0  # taste parameter
L = np.ones(S, dtype=float)  # start assuming labor supply in each region is 1
A = np.arange(1, S + 1, dtype=float)  # productivity A_i = 1, ..., 10
U = S + 1 - A # U_i is mirror image of productivity


# Define trade cost matrices
tau_high = np.full((S, S), 2.0) # tau_ij = 2 if i  j
np.fill_diagonal(tau_high, 1.0) # tau_ii = 1

tau_low = np.ones((S, S)) # free trade: tau_ij = 1 for all i, j


## ------ Inner loop: Solves for w keeping L constant  -------

# Fixed point solver on Y = lambda(Y) * Y
def solve_Y_lambdaY(tau, A, L, sigma, max_iter=1000, tol=1e-6):
    """
    Solves for equilibrium income vector Y using the fixed point Y = lambda(Y) * Y,
    where Y_i = income = w_i * L_i
    lambda_ji = share of income in i spent on goods from j
    """
    S = len(A)
    w = A.copy()     # Initial guess: assume autarky wages => w_i = A_i 
    Y = w * L        # Note the math is element-wise multiplication Y[i] = A[i] * L[i]

    for _ in range(max_iter):
        # Step 1: Compute price index P_j = [sum_i (p_ij)^(1 - sigma)]^{1/(1 - sigma)}
        P = np.array([
            sum((tau[i, j] * w[i] / A[i]) ** (1 - sigma) for i in range(S)) ** (1 / (1 - sigma))
            for j in range(S)
        ])

        # Step 2: Compute lambda_ij = (p_ij / P_j)^(1 - sigma)
        lambda_matrix = np.zeros((S, S))
        for j in range(S):  # Spending region
            for i in range(S):  # Producing region
                lambda_matrix[i, j] = ( (tau[i, j] * w[i] / A[i]) / P[j]) ** (1 - sigma)


        # Step 3: Update income Y using fixed point Y =  Y
        Y_new = lambda_matrix.T @ Y
        if np.max(np.abs(Y_new - Y)) < tol:
            break
        Y = Y_new.copy()
        w_new = Y / L  # Update wages
        w = w_new.copy()  # Update wages

    return w, P
    #return Y, lambda_matrix, P, w,  welfare

## ------ Outer loop: Solves for L keeping w constant  -------

def solve_spatial_equilibrium(tau, A, U, sigma, max_iter=1000, tol=1e-20):
    S = len(A)
    L = np.ones(S, dtype=float)

    for _ in range(max_iter):
        w, P = solve_Y_lambdaY(tau, A, L, sigma)
        W = (w / P) * U  # Indirect utility
        denominator = np.sum(W * L)
        L_new = (W * L / denominator) 
        #if np.max(np.abs(L_new - L)) < tol:
        if np.max(W) - np.mean(W) < tol:

            break
    
        L = L_new.copy()

    return L, w, P, W, w/P

## ----- RUN MODEL -----

# Low trade costs
L_low, w_low, P_low, W_low, rw_low= solve_spatial_equilibrium(tau_low, A, U, sigma)
# High trade costs  
L_high, w_high, P_high, W_high, rw_high = solve_spatial_equilibrium(tau_high, A, U, sigma)


# Output results
df_results = pd.DataFrame({
    "Region": np.arange(1, S + 1),
    "Productivity A": A,
    "Amenities U": U,
    "Labor (low)": L_low,
    "Labor (high)": L_high,
    "Wages (low)": w_low,
    "Wages (high)": w_high,
    "Prices (low)": P_low,
    "Prices (high)": P_high,
    "Real wage (low)": rw_low,
    "Real wage (high)": rw_high,
    "Welfare (low)": W_low,
    "Welfare (high)": W_high
    })

#print(df_results.round(3)) # display in a simple way

display(HTML(df_results.round(3).to_html())) # Display in a beautiful way 

# Plot: Population ratio between tau=1 and tau=2 ===
population_ratio = L_low / L_high

plt.figure(figsize=(8, 5))
plt.plot(A, population_ratio, marker='o', linestyle='-', color='orange')
plt.axhline(1, color='gray', linestyle='--')
plt.title("Population Change with Trade Cost Reduction (=1 / =2)")
plt.xlabel("Region Productivity (A_i)")
plt.ylabel("Population Ratio: L_i(=1) / L_i(=2)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Problem 2. The Basic Armington Model with Frictional Labor Mobility.

The structure of the spatial equilibrium condition changed so we now have a different outer loop, inner loop remains.


Indirect utility in location $i$ for agent $j$ is given by:

$$
W^j_i = f(w_i,P_i,U_i) \xi_i^j
$$

Where $\xi_i^j \sim F(\xi) = \exp(-\xi^{-\theta})$


Agent $i$ chooses locations so that  
$$
i^\star = \arg \max_i \{ W_i^j \} 
$$

This gives the fraction of workers that choose to live in $i$:
$$
\frac{L_i}{\bar{L}} = \pi_i = \frac{\left( f(w_i, P_i, U_i) \right)^\theta}{\sum_i \left( f(w_i, P_i, U_i) \right)^\theta}
= \frac{\left( \frac{w_i}{P_i} U_i \right)^\theta}{\sum_i \left( \frac{w_i}{P_i} U_i \right)^\theta}
$$


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML

# Set parameters
S = 10  # number of regions
sigma = 2.0  # CES elasticity
a_ij = 1.0  # taste parameter
L = np.ones(S, dtype=float)  # start assuming labor supply in each region is 1
A = np.arange(1, S + 1, dtype=float)  # productivity A_i = 1, ..., 10
U = S + 1 - A # U_i is mirror image of productivity
theta = 8.0

# Define trade cost matrices
tau_high = np.full((S, S), 2.0) # tau_ij = 2 if i  j
np.fill_diagonal(tau_high, 1.0) # tau_ii = 1

tau_low = np.ones((S, S)) # free trade: tau_ij = 1 for all i, j


## ------ Inner loop: Solves for w keeping L constant  -------

# Fixed point solver on Y = lambda(Y) * Y
def solve_Y_lambdaY(tau, A, L, sigma, max_iter=1000, tol=1e-6):
    """
    Solves for equilibrium income vector Y using the fixed point Y = lambda(Y) * Y,
    where Y_i = income = w_i * L_i
    lambda_ji = share of income in i spent on goods from j
    """
    S = len(A)
    w = np.ones(S, dtype=float)    
    Y = w * L        # Note the math is element-wise multiplication Y[i] = A[i] * L[i]

    for _ in range(max_iter):
        # Step 1: Compute price index P_j = [sum_i (p_ij)^(1 - sigma)]^{1/(1 - sigma)}
        P = np.array([
            sum((tau[i, j] * w[i] / A[i]) ** (1 - sigma) for i in range(S)) ** (1 / (1 - sigma))
            for j in range(S)
        ])

        # Step 2: Compute lambda_ij = (p_ij / P_j)^(1 - sigma)
        lambda_matrix = np.zeros((S, S))
        for j in range(S):  # Spending region
            for i in range(S):  # Producing region
                lambda_matrix[i, j] = ( (tau[i, j] * w[i] / A[i]) / P[j]) ** (1 - sigma)


        # Step 3: Update income Y using fixed point Y =  Y
        Y_new = lambda_matrix.T @ Y
        
        # Update wages **from the new Y**, before checking for convergence
        w_new = Y_new / L  # New wages after new income
        w = w_new.copy()
        
        #Update Y based on updated wages
        Y = w * L
        
        # Check convergence now
        if np.max(np.abs(Y_new - Y)) < tol:
            break

    return w, P

## ------ Equilibrium definition: Solves for L given w  -------

def solve_frechet(tau, A, U, sigma, theta, max_iter=1000, tol=1e-2):
    S = len(A)
    L = np.ones(S, dtype=float)

    for _ in range(max_iter):
        w, P = solve_Y_lambdaY(tau, A, L, sigma)
        W = (w / P) * U  # Indirect utility
        W_scaled = W / np.max(W)
        W_power = W_scaled ** theta
        pi = W_power / np.sum(W_power)
        L_new_aux = (pi * S)
        L_new = 0.05 * L_new_aux + 0.95 * L # If I followed the class expression some locations would have 0 labor, this does the same only slower
        L_new = (L_new / np.sum(L_new)) * S  # normalize to sum to S
        if np.max(np.abs(L_new - L)) < tol:
            break
        L = L_new.copy()

    return L, w, P, W, w/P

## ----- RUN MODEL :D -----

# Low trade costs
L_low, w_low, P_low, W_low, rw_low = solve_frechet(tau_low, A, U, sigma, theta)
# High trade costs  
L_high, w_high, P_high, W_high, rw_high = solve_frechet(tau_high, A, U, sigma, theta)


# Output results
df_results_2 = pd.DataFrame({
    "Region": np.arange(1, S + 1),
    "Productivity A": A,
    "Amenities U": U,
    "Labor (low)": L_low,
    "Labor (high)": L_high,
    "Wages (low)": w_low,
    "Wages (high)": w_high,
    "Prices (low)": P_low,
    "Prices (high)": P_high,
    "Real wage (low)": rw_low,
    "Real wage (high)": rw_high,
    "Welfare (low)": W_low,
    "Welfare (high)": W_high
    })


display(HTML(df_results_2.round(3).to_html())) # Display in a beautiful way 

# Plot: Population ratio between tau=1 and tau=2
population_ratio = L_low / L_high

plt.figure(figsize=(8, 5))
plt.plot(A, population_ratio, marker='o', linestyle='-', color='orange')
plt.axhline(1, color='gray', linestyle='--')
plt.title("Population Change with Trade Cost Reduction (=1 / =2)")
plt.xlabel("Region Productivity (A_i)")
plt.ylabel("Population Ratio: L_i(=1) / L_i(=2)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
